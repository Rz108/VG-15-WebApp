## <font color='#71a1e3'> __Model Deployment (Locally)__</font>
<hr style="height:1.5px;border-width:0;background-color: #4c8c77">


To deploy locally, docker command was use in command prompt
-   Because of the nature of two models on 1 link, model.config files was created to suppport that 
    - This is what model.condif file looks like 
    `model_config_list: {
  config: {
    name: "big",  
    base_path: "/models/img_classifier/big", 
    model_platform: "tensorflow"
  },
  config: {
    name: "small", 
    base_path: "/models/img_classifier/small",  
    model_platform: "tensorflow"
  }
}`
- `docker build -t tf-serving-img-classifier .`

To set up the contianer:
- `docker run -p 8501:8501 -p 8500:8500 tf-serving-img-classifier` was used to set up the container
